### Seção 01 - Trabalhando com Pandas SERIES

uma serie em Pandas e como uma coluna de uma tabela,ou seja ,uma matrz de apenas uma coluna. 

```python
#Utilizando  o comando pd.Series()
idades = pd.Series([15,20,25,30,40,50,60])
print(idades)
```

```python
#Retornar apenas os valores  da serie
idades.values
```

```python
#retornar apenas os indices dos registros
idades.index
```

```python
#Alterando os indices
indice_novo = ['Ana','Amanda','Jaqueline','Larissa','Roberto','João','Alan']
idades_list = [15,20,25,30,40,50,60]
nova_serie = pd.Series(idades_list,index= indice_novo)

print(nova_serie)
```

### Acessando informações de uma Serie

```python
print(nova_serie['Ana'])
```

### Seção 02:TRabalhando com DataFrame

Pandas DataFrame é uma estrutura de dados bidimensional,como uma matriz com múltiplas linhas e colunas.

```python
import pandas as pd 

dados = {
    'calorias':[420,380,390],
    'duracao':[50,40,45]

}
#carregando dados para um dataframe
df = pd.DataFrame(dados)

print(df)

#selecionando a coluna Caloria
print(df['calorias'])

#selecionando a coluna Caloria
print(df['duracao'])
```

```python
#Verificando apenas os valores do DataFrame
df.values
```

```python
#Verificando os indices do DataFrame
df.index
```
