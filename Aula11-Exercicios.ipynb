### Exercício 01 - Escreva um programa Python para desenhar uma linha com rótulo
adequado no eixo x, eixo y e um título.
Amostra de dados:

```python
# importar a biblioteca
import matplotlib.pyplot as plt

x = range(1,50)
y =[value * 3 for value in x]

print('Valores de x')
print(range(1,50))
print('Valores de y (Três vezes x)')
print(y)
plt.plot(x , y)
#adicionar um titulo ao grafico
plt.title('Draw a Line' , fontsize = 15 , c= 'red')

#adicionando titulo aos eixos
plt.xlabel('Valores de x')
plt.ylabel('Valores de y')
plt.xticks(fontsize = 8, c='red')
plt.yticks(fontsize =8, c='blue' ) 
```

### Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

#Retorna ( linspace) valores iggualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

```python
plt.plot(x,y, label = 'seno')
plt.plot (x,z,label='cosseno')
plt.plot (x,np.sqrt(x), label='raiz quadrada')
plt.legend()
```

### Matplotlib – Múltiplos subgráficos
Dois gráficos na vertical (linha)

```python
# o comando plt.subplot permite criar grupo de graficos
#plt.subplot (linhas,colunas,Posição de Elementos)

plt.subplot(2,1,1)    #primeiro garfico
plt.plot(x,y)        #conjunto de dados do primeiro graficos
plt.title('subplots')  

plt.subplot(2,1,2)     #Segundo gráfico
plt.plot(x,z)           #Conjunto de dados do segundo gráfico
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```

### Seaborn – Gráficos mais elaborados
Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos,
além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.
Importando a biblioteca
