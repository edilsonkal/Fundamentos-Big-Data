#### Grafico de linha data x data de venda

```python
#Cria  um novo dataframe utilizando as colunas data venda e valor venda
#e classifica o dataframe com base data venda
base = df[['Data Venda','ValorVenda']].sort_values('ValorVenda')
base = base.groupby('Data Venda').sum()
#base
x = base.index
y = base['ValorVenda']
#base.index
plt.plot(x, y)
```
