Importando as Bibliotecas

```python
import matplotlib.pyplot as plt #Permite a geração de graficos
import numpy as np #Permite trabalhar com matrizes e vetores 
```

### Primeiro Grafico com Matplotlib

```python
#Xraindo os vetores (cojunto de dados)
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

#Plotar o Grafico

plt.plot(x,y)
```

#### Modificando os tipos de marcadores

```python
plt.plot(x , y, marker='*')
```

#### Alterando as cores das linhas 

```python
plt.plot(x , y, marker ='*' , c='#F2D861')
```

mudando como os pontos

```python
plt.plot (
    x, y ,
    marker = '*',
    c = 'red' , 
    ds = 'steps',
    ls = '--'
)
```

```python
plt.plot (
    x, y ,
    marker = '*',
    c = 'red' , 
    ds = 'steps',
    ls = '--',
    mec = 'blue'
)
```

#### Matplotlib – adicionando outros elementos
Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outras
possibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando a
atenção do que realmente importa.


```python
#adicionando dados ao grafico
plt.plot(x , y)

#adicionando titulo ao eixo x
plt.xlabel('Valores x' , fontsize = 12)
plt.ylabel('Valores y' , fontsize = 12)
plt.title ('Titulo do Gráfico' , fontsize = 18 , c= 'purple')

#Altera as cores dos valores
plt.xticks(fontsize = 8, c='blue')
plt.yticks(fontsize = 8, c ='green')
```
